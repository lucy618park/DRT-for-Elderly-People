```python
df_od=df_old[df_old['출발지_행정동명']!=df_old['도착지_행정동명']]
df_od.drop(df_od[(df_od['출발지_시도명'] =='서울특별시')].index,inplace=True)
df_od=df_od.sort_values('인구 수',ascending=False)
df_od
```

```python
df_od['출발_도착']=df_od['출발지_행정동명']+"-"+df_od['도착지_행정동명'] #출발,도착 변수 합치기
df_od
```
